```python
#Gráfico de Jogos
Colordict = {'KobeBryant':'Black','JoeJohnson':'Blue','LeBronJames':'Green','CarmeloAnthony':'Gray','DwightHoward':'Yellow','ChrisBosh':'Red','ChrisPaul':'Purple','KevinDurant':'Violet','DerrickRose':'Brown','DwayneWade':'Orange'}
Markerdict = {'KobeBryant':'s','JoeJohnson':'^','LeBronJames':'v','CarmeloAnthony':'<','DwightHoward':'>','ChrisBosh':'*','ChrisPaul':'+','KevinDurant':'o','DerrickRose':'D','DwayneWade':'_'}
def gamesplot(playerlist):
    Colordict = {'KobeBryant':'Black','JoeJohnson':'Blue','LeBronJames':'Green','CarmeloAnthony':'Gray','DwightHoward':'Yellow','ChrisBosh':'Red','ChrisPaul':'Purple','KevinDurant':'Violet','DerrickRose':'Brown','DwayneWade':'Orange'}
    Markerdict = {'KobeBryant':'s','JoeJohnson':'^','LeBronJames':'v','CarmeloAnthony':'<','DwightHoward':'>','ChrisBosh':'*','ChrisPaul':'+','KevinDurant':'o','DerrickRose':'D','DwayneWade':'_'}
    for name in playerlist:
        plt.plot(Games[Pdict[name]], c=Colordict[name], ls='--', marker=Markerdict[name], ms=7, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)), Seasons, rotation='vertical')
    plt.show()
```

```python
#Corrigindo as inputs
def myplot(data, playerlist=Players):
    Colordict = {'KobeBryant':'Black','JoeJohnson':'Blue','LeBronJames':'Green','CarmeloAnthony':'Gray','DwightHoward':'Yellow','ChrisBosh':'Red','ChrisPaul':'Purple','KevinDurant':'Violet','DerrickRose':'Brown','DwayneWade':'Orange'}
    Markerdict = {'KobeBryant':'s','JoeJohnson':'^','LeBronJames':'v','CarmeloAnthony':'<','DwightHoward':'>','ChrisBosh':'*','ChrisPaul':'+','KevinDurant':'o','DerrickRose':'D','DwayneWade':'_'}
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Colordict[name], ls='--', marker=Markerdict[name], ms=7, label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)), Seasons, rotation='vertical')
    plt.show()
```

```python
#Gráfico de Pontos por Temporada
myplot(Points)
```

```python
myplot(Salary)
```

```python
def mynewplot(data, playerlist=Players):
    for name in playerlist:
        plt.plot(data[Pdict[name]], c=Colordict[name], marker=Markerdict[name], ms=7, ls='--', label=name)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)), Seasons, rotation='vertical')
    plt.show()
```

```python
mynewplot(Games)
```
