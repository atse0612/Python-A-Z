```python
#8 - Introdução ao Seaborn
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
%matplotlib inline
plt.rcParams['figure.figsize'] = 8,4
warnings.filterwarnings('ignore')
```

```python
#Distribuição
vis1 = sns.distplot(stats.InternetUsers)
```

```python
vis1 = sns.distplot(stats.InternetUsers,bins=30)
```

```python
#Boxplots:
vis2 = sns.boxplot(data=stats, x='IncomeGroup',y='BirthRate') #data=stats diz que nós quereoms usar o stats para o gráfico
```

```python
#9 - Visualizações com o Seaborn
vis1 = sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',fit_reg=False) #lm significa Linear Model
#x e y são as colunas do Dataframe que o Seaborn verifica. fit_reg mostra a reta média
```

```python
vis1 = sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',\
                  fit_reg=False, hue='IncomeGroup',size=10) #hue é para a cor, logo a legenda vai para a categoria Income Group
```

```python
#Pode-se observar que os países que possuem mais usuários de internet não possuem tantas crianças
```

```python
#-------KEYWORD ARGUMENTIS IN PYTHON (ADVANCED)
#marker size:
vis1 = sns.lmplot(data=stats, x='InternetUsers',y='BirthRate',\
                  fit_reg=False, hue='IncomeGroup',size=8,\
                 scatter_kws={'s':100})

#scatter_kws: Argumentos de keywords adicionais (necessário verificar a documentação). scatter_kws={'s':100} é o tamanho dos pontos
```
